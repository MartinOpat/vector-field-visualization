```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python

# Define grid size
n = 20  # Number of points in x and y
x = np.linspace(-10, 10, n)
y = np.linspace(-10, 10, n)

# Create a meshgrid
X, Y = np.meshgrid(x, y)

# Define the radial vector field
U = X  # X-component of the vector field
V = Y  # Y-component of the vector 

# Normalize the vector field
magnitude = np.sqrt(U**2 + V**2)
U_normalized = U / (magnitude + 1e-5)
V_normalized = V / (magnitude + 1e-5)

# Plot the vector field
plt.figure()
plt.quiver(X, Y, U_normalized, V_normalized, magnitude, cmap='viridis')
plt.title('Radial Vector Field')
plt.colorbar(label='Magnitude')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Define the circular/swirl vector field
U = -Y  # X-component of the vector field (rotation)
V = X   # Y-component of the vector field (rotation)

# Normalize the vector field
magnitude = np.sqrt(U**2 + V**2)
U_normalized = U / (magnitude + 1e-5)
V_normalized = V / (magnitude + 1e-5)

# Plot the vector field
plt.figure()
plt.quiver(X, Y, U_normalized, V_normalized, magnitude, cmap='plasma')
plt.title('Swirling Flow Field')
plt.colorbar(label='Magnitude')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Define the linear vector field with gradients
U = np.ones_like(X)  # Constant x-component (linear flow)
V = np.zeros_like(Y)  # No vertical component

# Create a gradient for the magnitude
magnitude = X + 10  # Magnitude increases from left to right

# Plot the vector field
plt.figure()
plt.quiver(X, Y, U, V, magnitude, cmap='cool')
plt.title('Linear Flow with Gradient')
plt.colorbar(label='Magnitude')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Define the turbulent (chaotic) vector field
U = np.random.randn(n, n)  # Random x-component of the vector field
V = np.random.randn(n, n)  # Random y-component of the vector field

# Calculate the magnitude of the vectors
magnitude = np.sqrt(U**2 + V**2)

# Plot the vector field
plt.figure()
plt.quiver(X, Y, U, V, magnitude, cmap='inferno')
plt.title('Turbulent Flow Field')
plt.colorbar(label='Magnitude')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import plotly.graph_objects as go
```

```python

# Define grid size
n = 10  # Number of points in each axis
x = np.linspace(-5, 5, n)
y = np.linspace(-5, 5, n)
z = np.linspace(-5, 5, n)

# Create a 3D meshgrid
X, Y, Z = np.meshgrid(x, y, z)

# Define the radial vector field in 3D
U = X  # X-component of the vector field
V = Y  # Y-component of the vector field
W = Z  # Z-component of the vector field

# Normalize the vector field (optional)
magnitude = np.sqrt(U**2 + V**2 + W**2)
U_normalized = U / (magnitude + 1e-5)
V_normalized = V / (magnitude + 1e-5)
W_normalized = W / (magnitude + 1e-5)

# Create 3D quiver plot
fig = go.Figure(data=go.Cone(
    x=X.flatten(),
    y=Y.flatten(),
    z=Z.flatten(),
    u=U_normalized.flatten(),
    v=V_normalized.flatten(),
    w=W_normalized.flatten(),
    colorscale='Blues',
    sizemode="absolute",
    sizeref=0.2,  # Scaling factor
    showscale=True))

# Customize layout
fig.update_layout(
    scene=dict(
        xaxis_title='X-axis',
        yaxis_title='Y-axis',
        zaxis_title='Z-axis'
    ),
    title="Interactive Radial Vector Field (3D)"
)

# Show interactive plot
fig.show()
```

```python
import ipywidgets as widgets
from IPython.display import display
```

```python


# Create a radial vector field in 3D
def generate_radial_vector_field(n):
    x = np.linspace(-5, 5, n)
    y = np.linspace(-5, 5, n)
    z = np.linspace(-5, 5, n)
    X, Y, Z = np.meshgrid(x, y, z)
    
    U = X  # X-component of the vector field
    V = Y  # Y-component of the vector field
    W = Z  # Z-component of the vector field
    
    magnitude = np.sqrt(U**2 + V**2 + W**2)
    U_normalized = U / (magnitude + 1e-5)
    V_normalized = V / (magnitude + 1e-5)
    W_normalized = W / (magnitude + 1e-5)
    
    return X, Y, Z, U_normalized, V_normalized, W_normalized

# Plot the 3D vector field
def plot_vector_field(scale, angle):
    X, Y, Z, U, V, W = generate_radial_vector_field(10)
    
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    # Apply scaling
    ax.quiver(X, Y, Z, U, V, W, length=scale, color='b')
    
    # Rotate the plot by the given angle
    ax.view_init(30, angle)
    
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')
    plt.title('Interactive Radial Vector Field (3D)')
    plt.show()

# Create sliders for interactivity
scale_slider = widgets.FloatSlider(min=0.1, max=2.0, step=0.1, value=0.5, description='Scale')
angle_slider = widgets.FloatSlider(min=0, max=360, step=1, value=30, description='Rotation Angle')

# Connect the sliders to the plot function
widgets.interact(plot_vector_field, scale=scale_slider, angle=angle_slider)

display(scale_slider, angle_slider)
```

# Color maps

```python
# Heatmap of the uniform 3d vector field
n = 10  # Number of points in each axis
x = np.linspace(-5, 5, n)
y = np.linspace(-5, 5, n)
z = np.linspace(-5, 5, n)

# Create a 3D meshgrid
X, Y, Z = np.meshgrid(x, y, z)

# Define the uniform vector field in 3D
U = np.ones_like(X)  # Constant x-component

# Create a 3D heatmap
fig = go.Figure(data=go.Volume(
    x=X.flatten(),
    y=Y.flatten(),
    z=Z.flatten(),
    value=U.flatten(),
    isomin=-1,
    isomax=1,
    opacity=0.1,  # Transparency
    surface_count=21,  # Number of isosurfaces
    colorscale='Viridis'
))

# Customize layout
fig.update_layout(
    scene=dict(
        xaxis_title='X-axis',
        yaxis_title='Y-axis',
        zaxis_title='Z-axis'
    ),
    title="Uniform 3D Vector Field"
)

# Show interactive plot
fig.show()

```

```python
# Use matplotlib to plot the uniform 3d vector field using heatmap
n = 50  # Increase the number of points in each axis for a more continuous appearance
x = np.linspace(-5, 5, n)
y = np.linspace(-5, 5, n)
z = np.linspace(-5, 5, n)

# Create a 3D meshgrid
X, Y, Z = np.meshgrid(x, y, z, indexing='ij')

# Define the uniform vector field in 3D
U = np.ones_like(X)  # Constant x-component

# Flatten the arrays for scatter plot
X_flat = X.flatten()
Y_flat = Y.flatten()
Z_flat = Z.flatten()
U_flat = U.flatten()

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create a 3D heatmap using scatter plot with decreased opacity
sc = ax.scatter(X_flat, Y_flat, Z_flat, c=U_flat, cmap='viridis', alpha=0.1)

# Add color bar to show the intensity scale
plt.colorbar(sc)

# Customize plot
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title('Uniform 3D Vector Field Heatmap')
plt.show()
```
